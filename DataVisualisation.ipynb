```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train.csv")
```

```python
df.sample(5)
```

```python
df.shape
```

```python
sns.histplot(df['Parch'],kde=False)
```

```python
sns.distplot(df['Age'],hist=False)
```

```python
plt.figure(figsize=(8,8))
sns.distplot(df['Age'])
```

```python
sns.relplot(x='Age', y='Fare', col='Pclass', hue='Sex', style='Sex', kind='line', data=df)
```

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x='Age', y='Fare', hue='Sex',data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.barplot(x='Sex', y='Survived', hue='Pclass', data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.stripplot(x='Sex', y='Age', hue='Sex', data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.swarmplot(x='Sex', y='Age', hue='Sex', data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.boxplot(x='Sex', y='Age', hue='Sex', data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.violinplot(x='Survived', y='Age', hue='Sex', data=df)
```

```python
sns.countplot(x='Survived', data=df)
```

```python
plt.figure(figsize=(8,8 ))
sns.pointplot(x='Sex', y='Survived', hue='Pclass', data=df)
```

```python
plt.subplots(figsize=(10,10))
sns.heatmap(df.select_dtypes(include=['number']).corr(), cmap='YlGnBu', annot=True, fmt='.2f')
```

```python
g = sns.FacetGrid(df, col="Sex", row="Survived", margin_titles=True, height=4)
g.map_dataframe(sns.histplot, x="Age", bins=20, kde=True, color="steelblue")

# Add labels and titles
g.set_axis_labels("Age", "Count")
g.set_titles(row_template="Survived = {row_name}", col_template="Sex = {col_name}")
plt.show()
```
